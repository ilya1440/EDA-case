Now one need to define the term similiraty betweeen two strings. When the length of two compared strings differs, additional internet resources suggest using Levenshtein distance as a proxy for the loss function. In essence, Levenshtein distance is the minimum number of single-character edits (insertions, deletions or substitutions) required to change one word into the other.

In our case two problems can occur. Firstly, 'CLAIM SPECILATY' column includes abbreviations. That means Levenshtein distance will be small for two abbreviations, however, the nature of these two abbreviations is different. Secondly, Levenshtein distance will be large if one string contains several words, including a keyword, and the other string contains only keyword, howerer, the nature of the two strings is same. In other words, ideally one would want the machine to identify the keyword in the string, however, I guess this kind of task is more difficult.

Given pair-wise Levenshtein distance, one need to cluster values. I belive some variation of k-means method would work here, however, I came across affinity propagation algorithm, which is easier to implemet in this particular case since we do not know the number of clusters. I did not go deep in the mechanics of this algorithm, though the basics of it is the following: there is a dymanic measure for each item, which calculates not only the extent of how the item is appropriate for a certain cluster, but also the extent of how item itself represents the charactrestics of a certain group. At the end of the day, the algorithm will produce some clusters and examplars corresponding to each cluster.

```python
# filter for rows with known claim specialty
claims = data['CLAIM_SPECIALTY'][data['CLAIM_SPECIALTY']!='UNKNOWN']

# convert pandas object to numpay array
strings = np.asarray(claims.unique())

# define pair-wise Levenshtein distance
lev_dist = -1*np.array([[distance.levenshtein(str1,str2) for str1 in strings] for str2 in strings])

# define algorithm with defualt parameters
ap = AffinityPropagation(affinity="precomputed", random_state=None)

# fit the model
ap.fit(lev_dist)

# inspect the results
for cluster in np.unique(ap.labels_):
    exemplar = strings[ap.cluster_centers_indices_[cluster]]
    cluster_values = np.unique(strings[np.nonzero(ap.labels_==cluster)])
    cluster_values_join = ", ".join(cluster_values)
    print(" --- [{}]: {}".format(exemplar, cluster_values_join))
```

Not that bad for the first touch of the problem. A significant part of the clusters indeed represent the same nature of claim. As expected, abbreviations were assigned into one cluster, though they represent different instances. Another problem is the case when two specialties have similar spelling but distinguish in nature, i.e. PSYCHOLOGY, CARDIOLOGY, LARYNGOLOGY.

For now I will not go further with this problem. Still I will make some adjustments. I will unite all clusters with abbreviations and cluster 'OTHERS' to represent one cluster with not very clear specialties.

```python
# populate dictionary clusters with specialties and exemplars
clusters = {}
for cluster in np.unique(ap.labels_):
    exemplar = strings[ap.cluster_centers_indices_[cluster]]
    cluster_values = np.unique(strings[np.nonzero(ap.labels_==cluster)])
    for specialty in cluster_values:
        if exemplar in ['CNA', 'SURGO', 'PAT', 'PEDER', 'ASC']:
            clusters[specialty] = 'OTHERS'
        else:
            clusters[specialty] = exemplar

# add 'UNKNOWN' specialty to clusters
clusters['UNKNOWN'] = 'UNKNOWN'
            
# create dataframe with specialties and clusters
clusters_df = pd.DataFrame(clusters.items(), columns=['CLAIM_SPECIALTY', 'CLUSTER'])
```

```python
# check that we did not lose any specialty throughout the procedure
if len(clusters_df)==data['CLAIM_SPECIALTY'].nunique():
    print('Everything is OK.')
else:
    print('Review the code.')
```

```python
# merge clusters_df with data
data_new = pd.merge(data, clusters_df, how='left', on='CLAIM_SPECIALTY')

# check whether cluster 'OTHERS' does not account for large proportion of the dataset
print("The proportion of rows with cluster 'OTHERS' is {}%.".format(round(len(data_new[data_new['CLUSTER']=='OTHERS'])*100/len(data_new), 2)))
```

# 2.2. Statistical Techniques

Popular technique to see whether categorical variables are related is Pearson’s Chi-Square Test, which essentially compares the observed and expected frequencies. The application of Pearson’s Chi-Square Test in this particular case seems problematic, because we have large number of subgroups for which frequency is calculated and large number of observations. That means some small divergence from expected frequencies for minor groups can lead to significant effect and rejection of the null hypothesis. What we want is a general measure of how variables are related. 

Now I will use Cramer's V to measure the absence of relation between categorical variables statistically. Cramer's V is a measure of association between two variables, which unlike correlation is in range [0,1] and like correlation is symmetric. Ideally, for categorical variables relation one need asymmetric measure, however, in this particular case the categorical variables seem to be independent, so we do need to find out what the exact realtion is, since there is probably no relation at all as the bar graps above show.

```python
# define fucntion to calculate Cramer V
def cramerV(x,y):
    # calculate observed frequencies
    confusion_matrix = pd.crosstab(x,y)

    # extract statistics
    chi2 = ss.chi2_contingency(confusion_matrix)[0]

    # calculate sample size
    n = confusion_matrix.sum().sum()

    # calculate Cramér's V
    phi2 = chi2/n
    r,k = confusion_matrix.shape
    phi2corr = max(0, phi2-((k-1)*(r-1))/(n-1))
    rcorr = r-((r-1)**2)/(n-1)
    kcorr = k-((k-1)**2)/(n-1)
    return np.sqrt(phi2corr/min((kcorr-1),(rcorr-1)))
```

```python
print("Cramer's V measure for 'PAYER' and 'SERVICE_CATEGORY' is {:f}.".format(cramerV(data_new['PAYER'], data_new['SERVICE_CATEGORY'] )))
```

```python
print("Cramer's V measure for 'PAYER' and 'CLAIM_SPECIALTY' is {:f}.".format(cramerV(data_new['PAYER'], data_new['CLUSTER'] )))
```

```python
print("Cramer's V measure for 'CLAIM_SPECIALTY' and 'SERVICE_CATEGORY' is {:f}.".format(cramerV(data_new['CLUSTER'], data_new['SERVICE_CATEGORY'] )))
```

As expected, the measure of association between 'PAYER' and 'SERVICE_CATEGORY' is very small. # PART 3. NUMERICAL AGGREGATION

I believe that one would be interested in the claims amount paid segmented by Payers, because these are Payers who finally pay for the medical services.

If we assume that the amount paid for a claim is mostly defined by Service Category, Claim Specialty and Payer, i.e. the effect of other factors that influence the amount paid is insignificant, we can assume that in general the nature of potential claim a Payer faces is identical. In other words, the distribution of potential amount paid for a claim is same each time for a Payer. This assumption can be valid, because previously we showed that 'PAYERS' and other columns are not related at least strongly.

Given the above assumption, one can futher assume that the sum of 'PAID_AMOUNT' by month by each Payer would represent a random variable distributed normally due to CLT, provided that the number of observations is sufficient. And for different payers the normal distribution will have different parameters.

Standard procedure to define whether a categorical variable affects a numerical variable is one-way ANOVA. The problem here is that we do not know how adjustments are distributed. For example, if for one Payer there are no adjustments and for the other adjustments cover the major part of claims, the calculation of means will have different nature for different Payers.

Though we could implement ANOVA on aggregated by month values, but since different Payers have different frequency occurence, the aggregated values will differ mostly due to this particular reason.

I suggest aggregating data by 'MONTH' and 'PAYER' column and checking whether aggreageted amounts paid are normally distributed for each Payer.

Nevertheless, firtly, lets see how 'AMOUNT_PAID' is distributed across different Payers.

```python
# plot the amount paid distribution in the whole dataset
# adjust range_x, so some extreme values do not show up
fig = px.histogram(
            data_frame=data_new['PAID_AMOUNT'],
            x='PAID_AMOUNT',
            range_x=[-10000,20000],
            title='Distribution of Amount Paid')
fig.show()
```

# PART 4. PREDICTION

Generally, I believe that the number of observations (31 months) is quite small to make any predictions for the future. If we try to fit some model to the data segmented by Payers, I doubt that the output will be reliable due to the limited number of observations and noise pertinent to individual Payers. Therefore, I suggest predicting the overall monhtly paid amount. The number of aggregated observations is still small, but noise is reduced. 

Actually prediction is not a proper word in this case, since as we saw earlier monthly aggregated amounts follow the normal distribution. That means a prediction for any month in the future is just a confidence interval for the mean, i.e. our model is just a constant plus noise.

Still I will look go through some standard procedures, which are generally applied to timeseries, though there is no much sense in them in this particular case.

```python
# convert 'MONTH' column to datetime format
data_new['MONTH'] = data_new['MONTH'].apply(lambda x: '{}-{}'.format(x[:4], x[4:]))
data_new['MONTH'] = pd.to_datetime(data_new['MONTH'])

# covert 'PAID_AMOUNT' to floats
data_new['PAID_AMOUNT'] = data_new['PAID_AMOUNT'].apply(lambda x: float(x))

# aggregate amounts paid by month and exclude 2 last months
pred = data_new.groupby(['MONTH'])['PAID_AMOUNT'].sum()[:-2]

# create function to test for stationarity
def DF_test(timeseries):
    # Dickey-Fuller test
    print('Results of Dickey-Fuller Test')
    df_test = adfuller(timeseries)
    df_output = pd.Series(df_test[0:4], index = ['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for key, value in df_test[4].items():
        df_output['Critical Value (%s)' %key] = value
    print(df_output)
```

```python
DF_test(pred)
```

As expected DF test rejected the hypothesis that timeseries have a unit root. Actually, we should have implemented the same procedure for higher orders of integration, but in our case threre is no theoretical reason to do it. I did not apply differencing to the timeseries, because they do not resemble any trend, seasonality and clear variance changes with time.  

```python
# plot acf
sm.graphics.tsa.plot_acf(pred, lags=10)
plt.show()
```

```python
# plot pacf
sm.graphics.tsa.plot_pacf(pred, lags=10)
plt.show()
```

As expected autocorrelation and partial autocorrelation coefficients are insignificant at any lag. Our model is ARIMA(0,0,0), i.e. a constant plus noise.

```python
# fit ARIMA(0,0,0) model
model = ARIMA(pred, order=(0, 0, 0), freq = 'MS')  
results_ARIMA = model.fit()
# plot projections
results_ARIMA.plot_predict(start = 1, end=38) 
plt.show()
```

All in all, we would obtain similar results, if we calculated just confidence interval for the mean using t-statistics.
